```python
# YouTube API Key (Replace with your actual API key)
API_KEY = "AIzaSyCXk7kNmpTMqTLJ3XWk-627S_7v5rNwmaQ"
youtube = build("youtube", "v3", developerKey=API_KEY)

# Function to fetch ALL comments from a video
def get_all_youtube_comments(video_id):
    comments = []
    next_page_token = None

    while True:  # Keep fetching until there are no more pages
        try:
            request = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                textFormat="plainText",
                maxResults=100,  # Fetch max comments per request
                pageToken=next_page_token  # Go to next page if available
            )
            response = request.execute()

            if "items" not in response:
                print(f"‚ùå No comments found for video: {video_id}")
                return []

            for item in response["items"]:
                comment = item["snippet"]["topLevelComment"]["snippet"]
                comments.append([
                    video_id,
                    comment["authorDisplayName"],
                    comment["textDisplay"],
                    comment["likeCount"],
                    comment["publishedAt"]
                ])

            # Check if there's another page of comments
            next_page_token = response.get("nextPageToken")

            if not next_page_token:  # If no more pages, break loop
                break

            print(f"‚úÖ Fetched {len(comments)} comments so far for {video_id}...")

            # Sleep to avoid API rate limits
            time.sleep(1)

        except Exception as e:
            print(f"‚ùå Error fetching comments for {video_id}: {e}")
            break  # Stop loop if API fails

    print(f"‚úÖ Successfully fetched {len(comments)} total comments for {video_id}")
    return comments

# ‚úÖ List of YouTube Video IDs
video_ids = ["rI78pAffkos", "RoGnlXQuJpk", "5Rc_Ka0AuSk", "clguUra-8QI", "bg4urpy-6LY"]

# Fetch comments for multiple videos
all_comments = []
for video_id in video_ids:
    print(f"üöÄ Fetching comments for video: {video_id}")
    comments_data = get_all_youtube_comments(video_id)
    all_comments.extend(comments_data)

# Save to CSV if comments were extracted
if len(all_comments) > 0:
    df_all_comments = pd.DataFrame(all_comments, columns=["Video_ID", "Author", "Comment", "Likes", "Published"])
    df_all_comments.to_csv("youtube_all_comments_bihar.csv", index=False)
    print("‚úÖ All comments saved to 'youtube_all_comments_bihar.csv'.")
else:
    print("‚ùå No comments were retrieved. Check API settings and video IDs.")
```
